```python
input_dir = 'Kaico_Images/June7'
# 保存するファルダ
output_dir = 'Kaico_Images/June7_tr'

# フォルダ内に保存されているファイル一覧を取得
image_list = list(pathlib.Path(input_dir).glob('**/*.jpg'))
```

```python
#image_list分、for分を回し、画像を切り取る
for i in range(len(image_list)):
    # 画像一枚をグレースケールで読み込み
    img = cv2.imread(str(image_list[i]))
    # trimming
    #トリミングしたい箇所を指定する
    img = img[5:610,220:970]
    # 保存先
    output_path = output_dir + '/' + image_list[i].name
    # 画像保存 
    cv2.imwrite(output_path, img)
```
